```python
from os import listdir
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
files = listdir('data')
print(files)
```

```python
with open('data/'+files[0]) as f:
    first_line = f.readline()
    print(first_line)
```

```python
data = np.genfromtxt('data/'+files[0], delimiter='\t')
```

```python
data[0]
```

```python
plt.plot(data[:,0])
plt.show()
```

```python
print(len(data)//2)
```

```python
fig, axs = plt.subplots(2, len(data)//2)
p=0
for i in range (2):
    for j in range(len(data)//2):
        if p < len(data):
            axs[i,j].plot(data[p])
            p+=1
        
```
