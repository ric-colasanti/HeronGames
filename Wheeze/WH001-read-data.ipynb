```python
fig, axs = plt.subplots(len(files)//2,2,figsize=(24, len(files)//2 * 6))
p=0
for i in range (2):
    for j in range(len(files)//2):
        if p < len(files):
            data = np.genfromtxt('data/'+files[p], delimiter='\t')
            print(files[p],data.shape)
            axs[j,i].plot(data[:,0])
            axs[j,i].set_title(files[p])
            p+=1
        
```
